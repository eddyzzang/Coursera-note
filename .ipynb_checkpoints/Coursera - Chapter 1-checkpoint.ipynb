## Cost Function
- We can measure the accuracy of our hypothesis function by using a cost function.
- This takes an average difference (actually a fancier version of an average) of all the results of the hypothesis with inputs from x's and the actual output y's.
- Also called "Squared error function", "Mean squared error"

Our objective is to get the best possible line. The best possible line will be such so that the average squared vertical distances of the scattered points from the line will be the least.

데이터 포인트를 지나는 최적의 선을 구하는 것이 목적. 데이터포인트와 최대한 가까운 선을 찾는 것.

#### θ1=1일 때 (Theta1은 기울기)
- Cost Function이 0임

#### θ1= 0.5일 때 
- Cost Function이 0.58로 올라감

#### 그래프를 그려보면
- 가장 작은 Cost Function을 찾는 것이 목표. 
- θ1=1 이 Global minimum이다.

### Contour plot
- 컨투어 플럿의 중심지점 즉, 밑으로 내려갈수록 선이 데이터포인트와 점점 가까워진다.

## Gradient Descent
- 가설이 얼마나 잘맞는지 확인하는 Cost Function을 알아보았으니 이제 가설의 파라미터를 측정해야한다. 이때 Gradient Descent가 쓰인다.

θ0는 X-axis, θ1는 Y-axis, J(θ0, θ1)는 Z-axis

어떤 지점에서 계속 내려가서 가장 낮은 지점을 찾는 것

The way we do this is by taking the derivative (the tangential line to a function) of our cost function. The slope of the tangent is the derivative at that point and it will give us a direction to move towards. We make steps down the cost function in the direction with the steepest descent. The size of each step is determined by the parameter α, which is called the learning rate.

Cost Function에서 미분를 해준다. Tangent의 기울기는 그 지점에서 미분해준 것이고 어느 방향으로 움직일지 알려준다. Cost Function 지점에서 가장 가파른 쪽으로 움직이며 움직이는 정도는 학습률인 파라미터 α에 의해 결정된다.

### Gradient descent algorithm
- 가장 낮은 점을 찾을 때 까지 계속 반복하며 파라미터를 계속 업데이트 해준다.

Tangent 기울기가 음수이면 θ1은 오른쪽으로 움직이고 기울기가 양수이면 왼쪽으로 움직인다

### Learning Rate

최소점으로 점점 다가갈수록 gradient descent는 자동으로 조금씩 움직인다. 그러므로 학습률을 줄일 필요가 없다.
